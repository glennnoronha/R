# Data Transformation with dplyr

```r
install.packages("nycflights13")
library(nycflights13)
library(tidyverse)
```

```r
glimpse(flights)
```

## Rows

### filter()<br>
display rows based on a column value without changing order (unlike arrange())

```r
# flights that had a departure delay > 120 mins
flights |>
  filter(dep_delay > 120) |>
  head()
```

All flights on January 1

```r
flights |>
  filter(month==1 &day==1)
```

Flights that departed in January or February<br>
Instead of having to use | and == you can use %in%

```r
flights |>
  filter(month %in% c(1,2))
```

Saving a new dataframe

```r
jan1 <- flights |>
  filter(month==1 & day==1)
glimpse(jan1)
```

### arrange()<br>
Selecting rows based on a value and arranging them in order

```r
flights |>
  arrange(year,month,day,dep_time)
```

```r
flights |>
  arrange(desc(dep_delay)) |>
  mutate(hours = dep_delay / 60)
```

### distinct() <br>
finds all unique rows of a data set (basically removes duplicate rows)

```r
flights |>
  distinct(origin,dest)
```

if you want to keep the other columns when using distinct() you can use .keep_all = TRUE

```r
flights |>
  distinct(origin,dest,.keep_all = TRUE)
```

find number of occurrences of a flight

```r
flights |>
  count(origin,dest,sort = TRUE) ### sort = TRUE equals descending order
```

## Columns

### mutate() <br>
used to add new columns based on existing columns

.before and .after can be used to move a mutated column accordingly <br>
you can target the index using it's position (int) or variable name

```r
flights |>
  mutate(
    gain = dep_delay - arr_delay,
    speed = distance / air_time * 60,
    .before = 1
  ) |>
    slice_head(n=5)
```

you can use .keep = "used" to only show the columns that were used in the calculation

```r
flights |>
  mutate(
    gain = dep_delay - arr_delay,
    speed = distance / air_time * 60,
    .keep = "used"
  ) |>
    slice_head(n=5)
```

### select()
select column values

```r
flights |>
  select(year:day)
```

```r
flights |>
  select(!year:hour)
```

select all columns that are integers

```r
flights |>
  select(where(is.integer))
```

starts_with("abc"): matches names that begin with “abc”. <br>

### rename()

```r
flights |>
  rename(YEAR = year, MONTH = month, DAY = day) |>
  slice_head(n=3)
```

### relocate()
by default it moves the variable to the front

```r
flights |>
  relocate(air_time) |>
  arrange(air_time) |>
  slice_head(n=5)
```
