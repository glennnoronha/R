###Install dplyr package and load it

```r
install.packages(('dplyr'))
library(dplyr)
```

Read the csv file

```r
data <- read.csv("charts.csv")
head(data)
```

Get only the past 5 years

```r
data <- data %>%
  mutate(year = as.integer(substr(date, 1, 4)))

last_5_years <- data %>%
  filter(year >= max(year)- 4)

range(last_5_years$year)
```
