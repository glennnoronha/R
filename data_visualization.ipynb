### Scatter Plot

aes() defines how data maps to axes
<br>
geom_point() draws the points using that mapping

```r
?aes
ggplot(
  data = penguins,
  mapping = aes(x=flipper_length_mm,y=body_mass_g)
  ) +
    geom_point()
```

using color to differentiate between species

```r
ggplot(
  data=penguins,
  mapping=aes(x=flipper_length_mm,y=body_mass_g,color=species)
) +
  geom_point()
```

adding a line of best fit using linear model method

```r
ggplot(
  data=penguins,
  mapping=aes(x=flipper_length_mm,y=body_mass_g,color=species)
) +
  geom_point() +
  geom_smooth(method="lm")
```

adding color only to the points themselves

```r
ggplot(
  penguins,
  aes(x=flipper_length_mm,y=body_mass_g)
) +
  geom_point(aes(color=species)) +
  geom_smooth(method = "lm")
```

Adding a different shape for each species

```r
ggplot(
  penguins,
  aes(x=flipper_length_mm,y=body_mass_g)
) +
  geom_point(aes(color=species, shape=species)) +
  geom_smooth(method = "lm")
```

Editting the labels to whatver we desire

```r
ggplot(
  penguins,
  aes(x=flipper_length_mm,y=body_mass_g)
) +
  geom_point(aes(color=species, shape=species)) +
  geom_smooth(method = "lm") +
  labs(
    title = "Body Mass (g) vs Flipper Length (mm)",
    subtitle = "Measurements for Adelie, Chinstrap and Gentoo Penguins",
    x = "Flipper Length (mm)", y = "Body Mass (g)",
    color = "Species", shape = "Species",
    caption="Data come from the palmerpenguins package"
  )
```

Recreating the same plot with piping

```r
penguins %>%
  ggplot(aes(x=flipper_length_mm,y=body_mass_g)) +
  geom_point(aes(color=species, shape=species)) +
  geom_smooth(method = "lm") +
  labs(
    title = "Body Mass (g) vs Flipper Length (mm)",
    subtitle = "Measurements for Adelie, Chinstrap and Gentoo Penguins",
    x = "Flipper Length (mm)", y = "Body Mass (g)",
    color = "Species", shape = "Species",
    caption="Data come from the palmerpenguins package"
  )
```

se=FALSE removes the display of the confidence band

```r
install.packages('patchwork')
library(patchwork)
p1 <- ggplot(
  data = penguins,
  mapping = aes(x = flipper_length_mm, y = body_mass_g, color = island)
) +
  geom_point() +
  geom_smooth()

p2 <- ggplot(
  data = penguins,
  mapping = aes(x = flipper_length_mm, y = body_mass_g, color = island)
) +
  geom_point() +
  geom_smooth(se = FALSE)

p1 + p2
```

### Bar Graph

geom_bar() creates a bar graph

```r
penguins %>%
  ggplot(aes(x = species)) +
  geom_bar()
```

Order the bars based on frequency from highest to lowest

```r
penguins %>%
  ggplot(aes(x = fct_infreq(species))) +
  geom_bar()
```

on the y-axis instead


```r
penguins %>%
  ggplot(aes(y = fct_infreq(species))) +
  geom_bar()
```

Differences between color and fill for bar charts <br>

fill is better for bar charts

```r
color_bar <- ggplot(penguins, aes(x = species)) +
  geom_bar(color = "red")

fill_bar <- ggplot(penguins, aes(x = species)) +
  geom_bar(fill = "red")

color_bar + fill_bar
```

### Histogram

use bin width to divide the bins by a custom amount. Here we do 200 g

```r
penguins %>%
  ggplot(aes(x = body_mass_g)) +
  geom_histogram(na.rm = TRUE, binwidth = 200)
```

### Density Plot

```r
penguins %>%
  ggplot(aes(x=body_mass_g)) +
  geom_density()
```
